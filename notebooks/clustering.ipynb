```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tslearn.utils import to_time_series, to_time_series_dataset

GROUP_LEVEL = 'LOCATION_ID'
```

```python
df = pd.read_csv('C:/Users/tijmen.vanetten/Documents/emobility-vanetten/data/processed/sessions_data.csv')
```

```python
df_daily = df.groupby([GROUP_LEVEL, 'DATE']).sum('ENERGY_DELIVERED_KWH')
df_locations = df_daily.groupby(GROUP_LEVEL)
```

```python
locations, time_series =zip(*list(df_locations['ENERGY_DELIVERED_KWH']))
ts_dataset = to_time_series_dataset(time_series)
```

```python
from tslearn.clustering import TimeSeriesKMeans
km = TimeSeriesKMeans(n_clusters=3, metric="dtw")
km.fit(ts_dataset)
```

```python
labels = km.fit_predict(ts_dataset)
```

```python
predictions = dict(zip(locations, labels))
predictions
```
