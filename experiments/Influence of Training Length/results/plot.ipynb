```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_rmse_mae(train_lengths, local_rmse_values, local_mae_values, global_rmse_values, global_mae_values):
    sns.set(style='whitegrid')  # Set the style to NeurIPS

    # Plot RMSE
    plt.figure(figsize=(8, 6))
    plt.plot(train_lengths, local_rmse_values, marker='o', label='Local')
    plt.plot(train_lengths, global_rmse_values, marker='o', label='Global')
    plt.plot(train_lengths, [local_rmse - global_rmse for local_rmse, global_rmse in zip(local_rmse_values, global_rmse_values)], marker='o', label='Difference')
    plt.title('Root Mean Squared Error (RMSE)')
    plt.xlabel('Training Length')
    plt.ylabel('RMSE Value')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()  # Adjust the layout for better spacing
    plt.show()

    # Plot MAE
    plt.figure(figsize=(8, 6))
    plt.plot(train_lengths, local_mae_values, marker='o', label='Local')
    plt.plot(train_lengths, global_mae_values, marker='o', label='Global')
    plt.plot(train_lengths, [local_mae - global_mae for local_mae, global_mae in zip(local_mae_values, global_mae_values)], marker='o', label='Difference')
    plt.title('Mean Absolute Error (MAE)')
    plt.xlabel('Training Length')
    plt.ylabel('MAE Value')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()  # Adjust the layout for better spacing
    plt.show()

# Example data
train_lengths = [600, 540, 480, 420, 360, 300, 240]
local_rmse_values = [93.56210805610786, 92.92908360508642, 92.92488980497548, 96.87339241084555, 96.32936892189275, 96.0675198382333, 101.5617085335908]
local_mae_values = [74.73536350160089, 73.78507081900854, 74.11557823750428, 78.31695232493851, 77.4104080583181, 77.22188148112976, 82.76908945848837]
global_rmse_values = [90.52604161556997, 87.17431322196437, 86.66482205159072, 85.85469783407324, 85.00319970685453, 87.80265996331559, 91.00775135006681]
global_mae_values = [71.8965167344335, 68.95355418972127, 68.15864489919137, 68.10268947984535, 67.09774195802486, 69.13813115828253, 72.00713264500904]

# Call the function to plot RMSE and MAE separately for local and global models, including the difference
plot_rmse_mae(train_lengths, local_rmse_values, local_mae_values, global_rmse_values, global_mae_values)
```
